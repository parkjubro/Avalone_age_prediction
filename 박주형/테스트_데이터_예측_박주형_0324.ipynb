```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd

# csv형식으로 된 데이터 파일을 읽어옵니다.
train_df = pd.read_csv('/Users/parkjubro/Desktop/Dacon/data/train.csv')
train_df.head()
```

```python
def make_label_map(dataframe):
    label_maps = {}
    for col in dataframe.columns:
        if dataframe[col].dtype == 'object':
            label_map = {'unknown':0}
            for i, key in enumerate(dataframe[col].unique()):
                label_map[key] = i+1
            label_maps[col] = label_map
    print(label_maps)
    return label_maps

def label_encoder(dataframe, label_map):
    for col in dataframe.columns:
        if dataframe[col].dtype=='object':
            dataframe[col] = dataframe[col].map(label_map[col])
            dataframe[col] = dataframe[col].fillna(label_map[col]['unknown'])
    return dataframe
```

```python
train_le = make_label_map(train_df[['Gender']])
gender_df = label_encoder(train_df[['Gender']], train_le)
train_df['Gender'] = gender_df[['Gender']]
```

```python
train_df.head()
```

```python
train_x = train_df.drop(['id','Target'], axis=1)
train_y = train_df.Target
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()

model.fit(train_x, train_y)
```

```python
test_df = pd.read_csv('/Users/parkjubro/Desktop/Dacon/data/test.csv')
```

```python
test_df.head()
```

```python
gender_df = label_encoder(test_df[['Gender']], train_le)
test_df['Gender'] = gender_df[['Gender']]

test_df.head()
```

```python
test_x = test_df.drop(['id'],axis=1)
```

```python
prediction = model.predict(test_x)
```

```python
submission = pd.read_csv('/Users/parkjubro/Desktop/Dacon/data/sample_submission.csv')
submission.head()
```

```python
submission['Target'] = prediction

submission
```
